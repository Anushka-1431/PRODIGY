```python
import tensorflow_hub as hub

hub_model = hub.load(
    "https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2"
)

print("Neural Style Transfer Model Loaded Successfully")
```

```python
import tensorflow as tf
import numpy as np
from PIL import Image

def load_image(path):
    img = Image.open(path)
    img = np.array(img).astype(np.float32) / 255.0

    img = tf.convert_to_tensor(img)

    if len(img.shape) == 3:
        img = img[tf.newaxis, :]

    return img

content_image = load_image("task_4.jpg")
style_image = load_image("task_5(1).jpeg")

print("Images Loaded Successfully")
```

```python
stylized_image = hub_model(
    tf.constant(content_image),
    tf.constant(style_image)
)[0]

print("Style Transfer Completed Successfully")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(content_image[0])
plt.title("Content Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(style_image[0])
plt.title("Style Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(stylized_image[0])
plt.title("Stylized Output")
plt.axis("off")

plt.show()
```

```python
from PIL import Image
import numpy as np

output_image = Image.fromarray(
    np.array(stylized_image[0] * 255, dtype=np.uint8)
)

output_image.save("stylized_output.png")

print("Stylized Image Saved Successfully")
```
