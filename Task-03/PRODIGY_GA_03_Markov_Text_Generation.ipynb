```python
text_model = markovify.Text(text)

print("Markov Model Created Successfully")
```

```python
for i in range(10):
    sentence = text_model.make_sentence(tries=100)

    if sentence:
        print(f"Sentence {i+1}:")
        print(sentence)
        print()
```

```python
print("Generated Paragraphs:\n")

for i in range(3):
    paragraph = text_model.make_short_sentence(
        250,
        tries=100
    )

    print(f"Paragraph {i+1}:")
    print(paragraph)
    print()
```
